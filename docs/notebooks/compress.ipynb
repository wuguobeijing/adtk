```python
import pandas as pd
from adtk.visualization import plot
from adtk.data import validate_series
s = pd.read_csv('./data/sensor_short1.csv', index_col="Time", parse_dates=True, squeeze=True)
s = validate_series(s)
```

```python
plot(s, ts_linewidth=2,ts_color='b', ts_markersize=5);
```
